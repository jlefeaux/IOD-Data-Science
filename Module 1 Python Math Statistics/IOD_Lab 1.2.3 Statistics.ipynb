````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):

    x = np.array(X)
    y = np.array(Y)
    N = len(x)

    '''
    Return m, b
    # ANSWER
    '''

    m = (N*(x*y).sum() - x.sum()*y.sum()) / (N*(x**2).sum() - x.sum()**2)
    b = (y.sum() - m*x.sum()) / N

    return m, b

# solution
# best_fit(X, Y)
m, b = best_fit(X, Y)
compare  = np.polyfit(X, Y, 1)

print(m, b)
print(compare)
````

Check following methods:

# Expected Output:
1. `stats.loggamma.rvs`: The gamma distribution is a continuous probability distribution. The shape is set with 2 parameters.  The exponential distribution, Erlang distribution, and chi-squared distribution are special cases of the gamma distribution.
    - `.rvs()` generates a set of samples drawn from the parent distribuion.  The number of samples drawn is set by the `size` parameter. Each random variable object type has it's own shape parameters.  For the loggamma distribution the shape parameter is a real number `c` > 0.
2. The probability plot is a graphical technique for assessing where or not a dataset follows a given distribution. The theoretical distribution is drawn as a straight line by plotting the CDF but scaling the y-axis so that the CDF appears as a line.  The data is plotted on the same axes.  The closer the data follow the distribution line, the better the fit is of the theoretical distribution to the data. 


````python
"""CODE DELETED. DO NOT TOUCH THIS TEXT. RUNNING THIS CELL WITH SHIFT-ENTER WILL DELETE THE GRAPH"""
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(211)
np.random.seed(0)
c = 5
x = stats.loggamma.rvs(c, size=500) + 5
# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# ANSWER
res1 = stats.probplot(x, plot=ax1)
ax1.set_title("Probability plot of a loggamma RV against a normal distribution")

# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution
ax2 = fig.add_subplot(212)
boxcox_lambda = 3
x_prime = x ** boxcox_lambda
res2 = stats.probplot(x_prime, plot=ax2)
ax2.set_title('Box-Cox transformation of the loggamma RV against a normal distribution')

plt.show()
````
