## 1. Extract the `country code`, `area code` and `local phone number` from the following phone numbers
**Hint**: Use groups with `()`. Assume they have the same form as shown.

```python
phonenumbers = '''+61 (02) 5443 5434
+64 (09) 888 7766
+1 (655) 432 4345'''

desired_output = [
    ('61', '02', '5443 5434'),
    ('64', '09', '888 7766'),
    ('1', '655', '432 4345')
]

country_code = r'\+(\d+)'
area_code = r'\((\d+)\)' 
phone_num = r'([\d\s]+)'
whitespace = r'\s*'
word_boundary = r'\b' # use this to not capture the newlines ('\n')

pattern = country_code + whitespace + area_code + whitespace + phone_num + word_boundary

output = re.findall(pattern, phonenumbers)

print(output == desired_output)

# assert output == desired_output, 'Not equal!'
```

## 2. Retrieve all the email addresses from the following text
Assume each email address has the form `id@domain.suffix`.

```python
text = '''Feel free to contact my managers p1@institutedata.com, p_2@domain2.com and p3@institutedata.com
for further details.'''

desired_output = [
    'p1@institutedata.com', 'p_2@domain2.com', 'p3@institutedata.com'
]

email_name = r'\w+'
at = r'@'
domain = r'\w+'
dot = r'\.'
suffix = r'\w+'

# Note: email_name, domain, and suffix are the value, but I repeated the pattern 
# with different variable names for clarity

pattern = email_name + at + domain + dot + suffix

output = re.findall(pattern, text)

print(output == desired_output)


# assert output == desired_output, 'Not equal!'
```

## 4. Filter the following tweet so that it contains only the message
In other words, remove URLs, RTs, CCs, hashtags, mentions, punctuations.

```python
tweet = 'Hello everyone! RT @MaskedAvenger: Did you see what I got up to earlier? http://abc.me/ref cc: @Mom_Judy #stayfit'

desired_output = 'Hello everyone Did you see what I got up to earlier'

url = r'http\S+\s*'
rt = r'RT|cc'
hashtag = r'#\S+'
mention = r'@\S+'
punctuation = r'[^\w\s]'
whitespace = r'\s+'
end_whitespace = r'\s+$'

def clean_tweet(tweet):
    tweet = re.sub(url, '', tweet)  # remove URLs
    tweet = re.sub(rt, '', tweet)  # remove RT and cc
    tweet = re.sub(hashtag, '', tweet)  # remove hashtags
    tweet = re.sub(mention, '', tweet)  # remove mentions
    tweet = re.sub(punctuation, '', tweet)  # remove punctuations
    tweet = re.sub(whitespace, ' ', tweet)  # remove internal extra whitespace
    tweet = re.sub(end_whitespace, '', tweet)  # remove extra whitespace at end
    return tweet

output = clean_tweet(tweet)
print(output)
print(desired_output)
print(output == desired_output)
# assert output == desired_output, 'Not equal!'
```
